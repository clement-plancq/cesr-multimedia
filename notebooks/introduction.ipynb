```python
# import Pillow library
from PIL import Image, ImageFilter

image = Image.open('../img/CPR_1130x400.jpg') # 'image' est un objet de la classe Image
#display(image)
```

Une fois en m√©moire on peut appliquer le filtre *blur* sur une partie de l‚Äôimage (voir la [doc](https://pillow.readthedocs.io/en/stable/reference/ImageFilter.html), toujours voir la doc)

```python
box = (185, 25, 335, 175) # on d√©finit la partie √† flouter avec ses coordonn√©es (left, top, right, bottom)
region = image.crop(box) # hop on d√©coupe la partie
blurred_region = region.filter(ImageFilter.BoxBlur(10)) # Blur‚ÄØpas Oasis
image.paste(blurred_region, box) # on copie la partie flout√©e

display(image)
```
