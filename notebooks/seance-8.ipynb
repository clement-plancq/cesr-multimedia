## Parcourir plusieurs fichiers dans un dossier

Utilisation du module (glob)[https://docs.python.org/3/library/glob.html]

```python
import glob
from PIL import Image
from IPython.display import display

# je cherche les fichiers .png dans le dossier ../img/
# équivalent à ls ../img/*.png sous Unix
png_files = glob.glob("../img/*.png")
for png_file in png_files:
    img = Image.open(png_file)
    display(img)
```
